```haskell
import IHaskell.Tables.Data
```

```haskell
tabular [ [ ["Bla", "blubb"]
          , ["Foo", "bar"  ]
          , ["Baz"]
          ]                , [["Muh"]], [["hiss"], ["sting", "qeek"]] ]
```

```haskell
tableWithLegend
    ("Stuff", "fuzz"          )
  [ (     17, "silver bullets")
  , (     42, "lives"         )
  , ( -32768, "overflows"     ) ]  :: Table [(Int, String)]
```

```haskell
tableWithLegend
    ( ("k","n"), "k^n" )
  [ ( ( k , n ),  k^n  )
  | k <- [1 .. 3 :: Int]
  , n <- [2 .. 4 :: Int] ]
```

```haskell
tableWithLegend
     ( ("k","n"), "k^n" )
 [ [ ( ( k , n ),  k^n  )
   | k <- [1 .. 3 :: Int] ]
 | n <- [2 .. 4 :: Int]     ]
```

```haskell
tableWithLegend
      ("n", "eⁿ")
    [ ( n, exp n)
    | n <- [0, 0.2 .. 3]] :: Table [(Double,Double)]
```

```haskell
tableWithLegend
      ("n", "ln (sin n)")
    [ ( n , log $ sin n)
    | n <- [-0.5, 0 .. 3]] :: Table [(Double,Double)]
```
